## Get additional info from csv

```python
midi_filename = 'midi_data/bwv104.6.mid'
#midi_filename = 'midi_data/pkgsc_azalea.mid'
csv_text = midi_utils.load_to_csv(midi_filename)
# Get beats-per-minute (default indicates if value was given in file or if midi-default was used)
value, default = midi_utils.get_bpm(csv_text)
print('bpm:', value)
# Find number of tracks in file and split tracks into seperate lists of csv-lines
track_dict = midi_utils.split_tracks(csv_text)
print('File has {} tracks'.format(len(track_dict.keys())-1)) # track 0 does not contain music
print('The track numbers are', track_dict.keys())
# Get number of midi-ticks per quarter note (default indicates if value was given in file or if midi-default was used)
# This info can be different for each track
for key in track_dict.keys():
    if key == '0':
        continue
    value, default = midi_utils.get_ticks_per_quarter(track_dict[key])
    print('Track {0} has {1} midi-ticks per quarter note'.format(key, value))
```
