## Extract numpy array from csv
Resulting shape: array of tracks which each contain a list of notes, where each note is array of 2 values: pitch, length (in midi-ticks)

```python
midi_filename = 'midi_data/bwv104.6.mid'
reduced_midi_filename = 'midi_data/test_single_track.mid'
midi_utils.reduce_to_single_track(midi_filename, reduced_midi_filename, '1')
# operate on single-track file because modified track will not fit to others (due to changed note lengths)
csv_text = midi_utils.load_to_csv(reduced_midi_filename)
# Split into tracks
track_dict = midi_utils.split_tracks(csv_text)
track_nr = '1'
# Generating numpy array with notes which will be modified later
numpy_notes = midi_utils.midi_track_to_numpy(track_dict[track_nr])
print(numpy_notes)
```

#### Modify resulting numpy array (all notes same length -> two quarters long)

```python
# make all notes the same length (2*quarter note)
for note in numpy_notes:
    note[1] = 2048
```

#### Convert back to a track with midi commands

```python
new_track = midi_utils.numpy_to_midi_track(numpy_notes, 1, 'Modified')
print(new_track)
```

#### Replace track1 with modified version and write new midi file to disc

```python
track_dict['1'] = new_track
modified_midi_filename = 'midi_data/test_modified_track.mid'
modified_csv_list = midi_utils.track_dict_to_csv(track_dict)
midi_utils.write_to_midi(modified_csv_list, modified_midi_filename)
```

```python
# play original single-track
pygame.init()
pygame.mixer.music.load(reduced_midi_filename)
pygame.mixer.music.play()
```

```python
pygame.mixer.music.stop()
```

```python
# play modified
pygame.init()
pygame.mixer.music.load(modified_midi_filename)
pygame.mixer.music.play()
```

```python
pygame.mixer.music.stop()
```

## Finding the lead track of midi file

```python
midi_filename = 'midi_data/bwv104.6.mid'
reduced_midi_filename = 'midi_data/lead_track.mid'
csv_text = midi_utils.load_to_csv(midi_filename)

# Split tracks into seperate lists of csv-lines
track_dict = midi_utils.split_tracks(csv_text)
```

```python
# Setting applied policies for lead track extraction
# If multiple policies (except random) are selected the track with the highest score is chosen
# If there is a tie, random selection between the tied tracks

# if random is selected, all other policies are ignored
random = False
# largest difference between highest and lowest note
largest_range = True
# largest number of different notes
most_variation = True
# largest number of different note lengths (indicates most complex rhythm)
most_rhythm = True

lead_nr = midi_utils.find_lead_track(track_dict, random=random, largest_range=largest_range, 
                                     variation=most_variation, rhythm=most_rhythm)

print('Choosing track nr:', lead_nr)
print(track_dict[lead_nr][:2])
```

```python
# Write file that only contains single track
midi_utils.reduce_to_single_track(midi_filename, reduced_midi_filename, lead_nr)
# Transform lead track to numpy
numpy_lead = midi_utils.midi_track_to_numpy(track_dict[lead_nr])
```
